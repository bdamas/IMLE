```python
from imle import Imle

d, D = 1, 1
imle_obj = Imle(d, D, Psi0=[0.1]*D, sigma0=1.0, p0=0.2, multiValueSignificance=0.9)

imle_obj.param
```

```python
NTRAIN = 2000
NOISE_Z = 0.02
NOISE_X = 0.1

z = arange(0.0, NTRAIN*0.1, 0.1) % (4*pi) + uniform(size=NTRAIN) * NOISE_Z
x = cos(z) + normal(size=NTRAIN) * NOISE_X

z = z.reshape(-1, 1)
x = x.reshape(-1, 1)

imle_obj.update(z, x)

print imle_obj.number_of_experts
```

```python
NQUERY = 1000

z = (uniform(size=NQUERY) * 4 * pi).reshape(-1, 1)
x = imle_obj.predict(z)
plot(z, x, 'o')
```

```python
error = abs(cos(z) - x)
plot(error)
mean(error), std(error)
```

```python
res = imle_obj.predict(z, jacobian=True)
pred = res['prediction']
jacobian = res['jacobian']

jacobian_err = abs(jacobian[:, :, 0] - (-sin(z)))
plot(jacobian_err)
mean(jacobian_err), std(jacobian_err)
```

```python
NQUERY = 1000

x = (uniform(size=NQUERY) * 2.0 - 1.0).reshape(-1, 1)
inv_x = 1.0 / x

inv_sol = imle_obj.predict_inverse(x)

hist([len(s) for s in inv_sol])
```
