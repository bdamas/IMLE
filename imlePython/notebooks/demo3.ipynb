```python
from imle import Imle
```

```python
d, D = 1, 1
imle_obj = Imle(d, D, Psi0=[0.1]*D, sigma0=1.0, p0=0.3)
```

```python
imle_obj.param
```

```python
NTRAIN = 2000
NOISE_Z = 0.02
NOISE_X = 0.1

zTrain = 0.0
z, x = [0.0], [0.0]

for _ in range(NTRAIN):
    zTrain += uniform() * NOISE_Z
    
    if (zTrain > 0.0):
        z[0] = zTrain
        x[0] = cos(z[0]) + NOISE_X * normal()
    else:
        z[0] = zTrain + 4 * pi
        x[0] = cos(z[0]) + 4.0 + NOISE_X * normal()
    
    imle_obj.update(z,x)
    
    if (zTrain > 4 * pi):
        zTrain = - 4 * pi
    else:
        zTrain += 0.1
        
imle_obj.number_of_experts
```

```python
z = linspace(0, 4 * pi, 100).reshape(-1, 1)
pred = imle_obj.predict(z, multiple=True)

plot(cos(z), 'k', linewidth=5)
plot(cos(z) + 4., 'k', linewidth=5)

plot(pred[:, 0], 'o')
plot(pred[:, 1], 'o')
```
